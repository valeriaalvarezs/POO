Vamos a comparar estadisticamente una lista de tickers

```python
#vamos a descargar los tickers, con las fechas dadas
fecha_inicial="2021-12-31"
fecha_final="2023-10-05"
```

```python
# descargamos la informacion de femsa con las fechas anteriores
femsa= yf.download("FEMSAUBD.MX",fecha_inicial,fecha_final)
femsa
```

```python
# descargamos la informacion de herdez con las fechas anteriores
herdez= yf.download("HERDEZ.MX",fecha_inicial,fecha_final)
herdez
```

```python
# descargamos la informacion de cuervo con las fechas anteriores
cuervo=yf.download("CUERVO.MX",fecha_inicial,fecha_final)
cuervo
```

```python
# descargamos la informacion de alsea con las fechas anteriores
alsea=yf.download("ALSEA.MX",fecha_inicial,fecha_final)
alsea
```

```python
# descargamos la informacion de ac con las fechas anteriores
ac=yf.download("AC.MX",fecha_inicial,fecha_final)
ac
```

```python
# hacemos un data frame con los tickers descargados
tickers=["FEMSAUBD.MX","HERDEZ.MX","CUERVO.MX","ALSEA.MX","AC.MX"]

datos = yf.download(tickers,fecha_inicial,fecha_final)["Close"]
datos
```

```python
# se describen los datos de los precios
datos.describe()
```

```python
# desviacion estandar de las empresas
datos.std()
```

```python
# varianza de los datos
varianza=datos.std()**2
```

```python
# se muestra la varianza
varianza=datos.var()
varianza
```

```python
# ahora con pct change se sacan los rendimientos de los datos
rendimientos= datos.pct_change()
rendimientos
```

```python
# se describen los rendimientos
rendimientos.describe()
```

```python
# varianza de los rendimientos
rendimientos.var()*100
```

```python
# covarianza de los rendimientos
rendimientos.cov()
```

```python
# es otra forma de sacar los rendimientos con formula
herdez["Rendimientos"]=(herdez["Close"]/herdez["Close"].shift(1))-1
herdez=herdez[["Close","Rendimientos"]]
herdez
```

```python
# grafica de los rendimientos de herdez
herdez["Rendimientos"].hist(bins=30)
```
